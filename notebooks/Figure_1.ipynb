```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
from matplotlib import offsetbox
from scipy.io import loadmat

from matplotlib.legend_handler import HandlerLine2D
plt.rc('font', family='serif')
plt.rc('text', usetex=True)
```

```python

w0 = 1
w02 = w0**2
d0 = 0.1
r0=1.5
x = np.linspace(-2, 2, 1000)

fig = plt.figure(figsize=(17, 6))
ax = fig.add_subplot(1, 2, 1)

r = 1
d = 0 
U = 0.5*r*w02*x**2 + 0.25*d*x**4
ax.plot(x, U,'-',linewidth=2.5,label='$r=1,~\delta = 0$')

r = 1
d = d0 
U = 0.5*r*w02*x**2 + 0.25*d*x**4
ax.plot(x, U,'--',linewidth=2.5,label='$r=1,~\delta > 0$')

r = -1
d = 20*d0 
U = 0.5*r*w02*x**2 + 0.25*d*x**4
ax.plot(x, U,'-.',linewidth=2.5,label='$r=-1,~\delta > 0$')

ax.set_ylim(-0.5,2)
ax.set_xlim(-2,2)
ax.legend(prop={'size':23}, framealpha=0.0)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)

ax.set_ylabel('$U(x)$', fontsize=30)
ax.set_xlabel('$x$', fontsize=30)
ax.set_title('$a)$',fontsize=40)

ax = fig.add_subplot(1, 2, 2)

r = 1
d = 0 
U = r*w02*x + d*x**3
ax.plot(x, U,'-',linewidth=2.5,label='$r=1,~\delta = 0$')

r = 1
d = d0 
U = r*w02*x + d*x**3
ax.plot(x, U,'--',linewidth=2.5,label='$r=1,~\delta > 0$')

r = -1
d = 20*d0 
U = r*w02*x + d*x**3
ax.plot(x, U,'-.',linewidth=2.5,label='$r=-1,~\delta > 0$')

ax.set_ylim(-4,4)
ax.set_xlim(-2,2)
ax.legend(prop={'size':23}, framealpha=0.0)

ax.set_yticks(np.arange(-4, 4.1, 1))
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)

ax.set_ylabel(r'$\frac{dU(x)}{dx}$', fontsize=30)
ax.set_xlabel('$x$', fontsize=30)
ax.set_title('$b)$',fontsize=40)



fig.tight_layout()

plt.draw()
plt.savefig('figure_1.eps',dpi=1000)
plt.show()
```
